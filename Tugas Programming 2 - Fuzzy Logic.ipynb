```python
import pandas as pd
import numpy as np

class Bengkel :
  ID = 0
  Servis = 0
  Harga = 0
  Kelayakan = 0

class SetServis : 
  Buruk = 0
  Cukup = 0
  Baik = 0
  SangatBaik = 0


class SetHarga :
  Murah = 0
  Menengah = 0
  Mahal = 0

class FuzzySet :
  Servis = SetServis()
  Harga = SetHarga()

class SetNK :
  Rendah = 0
  Tinggi = 0

def ReadBengkel(array, df) :
  for i in range(0,100) :
    temp = Bengkel()
    temp.ID, temp.Servis, temp.Harga = df["id"][i], df["servis"][i], df["harga"][i]
    array.append(temp)

def SetHargaValue(harga) :
  temp = SetHarga
  if harga >= 0 and harga <= 2 :
    temp.Murah = 1
    temp.Menengah = 0
    temp.Mahal = 0
  
  elif harga > 2 and harga < 5 :
    temp.Murah = (5 - harga) / 3
    temp.Menengah = (harga - 2) / 3
    temp.Mahal = 0

  elif harga >= 5 and harga <= 6 :
    temp.Murah = 0
    temp.Menengah = 1
    temp.Mahal = 0

  elif harga > 6 and harga < 8 :
    temp.Murah = 0
    temp.Menengah = (8 - harga) / 3
    temp.Mahal = (harga - 5) / 3

  elif harga >= 8 and harga <= 10 :
    temp.Murah = 0
    temp.Menengah = 0
    temp.Mahal = 1

  return temp

def SetServisValue(servis) :
  temp = SetServis()
  if servis >= 0 and servis <= 10 :
    temp.Buruk = 1
    temp.Cukup = 0
    temp.Baik = 0
    temp.SangatBaik = 0

  elif servis > 10 and servis < 35 :
    temp.Buruk = (35 - servis) / 20
    temp.Cukup = (servis - 10) / 20
    temp.Baik = 0
    temp.SangatBaik = 0

  elif servis >= 35 and servis <= 45 :
    temp.Buruk = 0
    temp.Cukup = 1
    temp.Baik = 0
    temp.SangatBaik = 0

  elif servis > 45 and servis < 65 :
    temp.Buruk = 0
    temp.Cukup = (65 - servis) / 20
    temp.Baik = (servis - 45) / 20
    temp.SangatBaik = 0

  elif servis >= 65 and servis <= 75 :
    temp.Buruk = 0
    temp.Cukup = 0
    temp.Baik = 1
    temp.SangatBaik = 0

  elif servis > 75 and servis < 95 :
    temp.Buruk = 0
    temp.Cukup = 0
    temp.Baik = (95 - servis) / 20
    temp.SangatBaik = (servis - 75) / 20

  elif servis >= 95 and servis <= 100 :
    temp.Buruk = 0
    temp.Cukup = 0
    temp.Baik = 0
    temp.SangatBaik = 1

  return temp

def fuzzification(data) :
  array = []
  for i in range(0,100) :
    temp = FuzzySet()
    temp.Servis = SetServisValue(data[i].Servis)
    temp.Harga = SetHargaValue(data[i].Harga)
    array.append(temp)
  return array

def FuzzyRules(servis, harga) :
  if servis == "Buruk" :
    return "Rendah"
  elif servis == "Cukup" and harga == "Murah" :
    return "Tinggi"
  elif servis == "Cukup" and harga == "Menengah" :
    return "Rendah"
  elif servis == "Cukup" and harga == "Mahal" :
    return "Rendah"
  elif servis == "Baik" and harga == "Murah" :
    return "Tinggi"
  elif servis == "Baik" and harga == "Menengah" :
    return "Tinggi"
  elif servis == "Baik" and harga == "Mahal" :
    return "Rendah"
  elif servis == "Sangat Baik" and harga == "Murah" :
    return "Tinggi"
  elif servis == "Sangat Baik" and harga == "Menengah" :
    return "Tinggi"
  elif servis == "Sangat Baik" and harga == "Mahal" :
    return "Tinggi"
  return "Error"

def Inferensi(fuzzyset) :
  NK = SetNK()
  if fuzzyset.Servis.Buruk > 0 :
    if fuzzyset.Harga.Murah > 0 :
      Konjungsi = min(fuzzyset.Servis.Buruk,fuzzyset.Harga.Murah) #konjungsi
      if FuzzyRules("Buruk", "Murah") == "Rendah" :
        NK.Rendah = max(NK.Rendah, Konjungsi) #disjungsi
      elif FuzzyRules("Buruk", "Murah") == "Tinggi" :
        NK.Tinggi = max(NK.Tinggi, Konjungsi) #disjungsi
    elif fuzzyset.Harga.Menengah > 0 :
      Konjungsi = min(fuzzyset.Servis.Buruk,fuzzyset.Harga.Menengah) #konjungsi
      if FuzzyRules("Buruk", "Menengah") == "Rendah" :
        NK.Rendah = max(NK.Rendah, Konjungsi) #disjungsi
      elif FuzzyRules("Buruk", "Menengah") == "Tinggi" :
        NK.Tinggi = max(NK.Tinggi, Konjungsi) #disjungsi
    elif fuzzyset.Harga.Mahal > 0 :
      Konjungsi = min(fuzzyset.Servis.Buruk,fuzzyset.Harga.Mahal) #konjungsi
      if FuzzyRules("Buruk", "Mahal") == "Rendah" :
        NK.Rendah = max(NK.Rendah, Konjungsi) #disjungsi
      elif FuzzyRules("Buruk", "Mahal") == "Tinggi" :
        NK.Tinggi = max(NK.Tinggi, Konjungsi) #disjungsi

  if fuzzyset.Servis.Cukup > 0 :
    if fuzzyset.Harga.Murah > 0 :
      Konjungsi = min(fuzzyset.Servis.Cukup,fuzzyset.Harga.Murah) #konjungsi
      if FuzzyRules("Cukup", "Murah") == "Rendah" :
        NK.Rendah = max(NK.Rendah, Konjungsi) #disjungsi
      elif FuzzyRules("Cukup", "Murah") == "Tinggi" :
        NK.Tinggi = max(NK.Tinggi, Konjungsi) #disjungsi
    elif fuzzyset.Harga.Menengah > 0 :
      Konjungsi = min(fuzzyset.Servis.Cukup,fuzzyset.Harga.Menengah) #konjungsi
      if FuzzyRules("Cukup", "Menengah") == "Rendah" :
        NK.Rendah = max(NK.Rendah, Konjungsi) #disjungsi
      elif FuzzyRules("Cukup", "Menengah") == "Tinggi" :
        NK.Tinggi = max(NK.Tinggi, Konjungsi) #disjungsi
    elif fuzzyset.Harga.Mahal > 0 :
      Konjungsi = min(fuzzyset.Servis.Cukup,fuzzyset.Harga.Mahal) #konjungsi
      if FuzzyRules("Cukup", "Mahal") == "Rendah" :
        NK.Rendah = max(NK.Rendah, Konjungsi) #disjungsi
      elif FuzzyRules("Cukup", "Mahal") == "Tinggi" :
        NK.Tinggi = max(NK.Tinggi, Konjungsi) #disjungsi

  if fuzzyset.Servis.Baik > 0 :
    if fuzzyset.Harga.Murah > 0 :
      Konjungsi = min(fuzzyset.Servis.Baik,fuzzyset.Harga.Murah) #konjungsi
      if FuzzyRules("Baik", "Murah") == "Rendah" :
        NK.Rendah = max(NK.Rendah, Konjungsi) #disjungsi
      elif FuzzyRules("Baik", "Murah") == "Tinggi" :
        NK.Tinggi = max(NK.Tinggi, Konjungsi) #disjungsi
    elif fuzzyset.Harga.Menengah > 0 :
      Konjungsi = min(fuzzyset.Servis.Baik,fuzzyset.Harga.Menengah) #konjungsi
      if FuzzyRules("Baik", "Menengah") == "Rendah" :
        NK.Rendah = max(NK.Rendah, Konjungsi) #disjungsi
      elif FuzzyRules("Baik", "Menengah") == "Tinggi" :
        NK.Tinggi = max(NK.Tinggi, Konjungsi) #disjungsi
    elif fuzzyset.Harga.Mahal > 0 :
      Konjungsi = min(fuzzyset.Servis.Baik,fuzzyset.Harga.Mahal) #konjungsi
      if FuzzyRules("Baik", "Mahal") == "Rendah" :
        NK.Rendah = max(NK.Rendah, Konjungsi) #disjungsi
      elif FuzzyRules("Baik", "Mahal") == "Tinggi" :
        NK.Tinggi = max(NK.Tinggi, Konjungsi) #disjungsi

  if fuzzyset.Servis.SangatBaik > 0 :
    if fuzzyset.Harga.Murah > 0 :
      Konjungsi = min(fuzzyset.Servis.SangatBaik,fuzzyset.Harga.Murah) #konjungsi
      if FuzzyRules("Sangat Baik", "Murah") == "Rendah" :
        NK.Rendah = max(NK.Rendah, Konjungsi) #disjungsi
      elif FuzzyRules("Sangat Baik", "Murah") == "Tinggi" :
        NK.Tinggi = max(NK.Tinggi, Konjungsi) #disjungsi
    elif fuzzyset.Harga.Menengah > 0 :
      Konjungsi = min(fuzzyset.Servis.SangatBaik,fuzzyset.Harga.Menengah) #konjungsi
      if FuzzyRules("Sangat Baik", "Menengah") == "Rendah" :
        NK.Rendah = max(NK.Rendah, Konjungsi) #disjungsi
      elif FuzzyRules("Sangat Baik", "Menengah") == "Tinggi" :
        NK.Tinggi = max(NK.Tinggi, Konjungsi) #disjungsi
    elif fuzzyset.Harga.Mahal > 0 :
      Konjungsi = min(fuzzyset.Servis.SangatBaik,fuzzyset.Harga.Mahal) #konjungsi
      if FuzzyRules("Sangat Baik", "Mahal") == "Rendah" :
        NK.Rendah = max(NK.Rendah, Konjungsi) #disjungsi
      elif FuzzyRules("Sangat Baik", "Mahal") == "Tinggi" :
        NK.Tinggi = max(NK.Tinggi, Konjungsi) #disjungsi
  return NK

def defuzzification(Rendah, Tinggi) :
  pembilang = 0
  penyebut = 0
  for i in range(5,65,5) :
    pembilang += i*Rendah
    penyebut += Rendah

  for i in range(65,80,5) :
    batasRendah = (i - 50)/30
    batasTinggi = (80 - i)/30
    if batasRendah > batasTinggi :
      if Rendah > batasRendah :
        pembilang += i*batasRendah
        penyebut += batasRendah
      elif batasRendah > Rendah :
        if Rendah > batasTinggi :
          pembilang += i*Rendah
          penyebut += Rendah
        elif batasTinggi > Rendah :
          if Tinggi > batasTinggi :
            pembilang += i*batasTinggi
            penyebut += batasTinggi
          elif batasTinggi > Tinggi :
            pembilang += i*Tinggi
            penyebut += Tinggi
    elif batasTinggi > batasRendah :
      if Tinggi > batasTinggi :
        pembilang += i*batasTinggi
        penyebut += batasTinggi
      elif batasTinggi > Tinggi :
        if Tinggi > batasRendah :
          pembilang += i*Tinggi
          penyebut += Tinggi
        elif batasRendah > Tinggi :
          if Rendah > batasRendah :
            pembilang += i*batasRendah 
            penyebut += batasRendah
          elif batasRendah > Rendah :
            pembilang += i*Rendah
            penyebut += Rendah

  for i in range(80,101,5) :
      pembilang += i*Tinggi
      penyebut += Tinggi


  Hasil = pembilang / penyebut
  return Hasil

def insertionSort(arr):
  key = Bengkel()
  for i in range(1, len(arr)):
      key = arr[i]
      j = i-1
      while j >= 0 and key.Kelayakan > arr[j].Kelayakan :
              arr[j + 1] = arr[j]
              j -= 1
      arr[j + 1] = key
```

```python
url = 'https://raw.githubusercontent.com/nicvalentino/TugasProgramming2AI/master/bengkel.csv'
df = pd.read_csv(url)

tabBengkel = []
ReadBengkel(tabBengkel,df)

fuzzyset = []
fuzzyset = fuzzification(tabBengkel) #Fuzzification

arrayNK =  []
for i in range(0,100) : #Inferensi
  arrayNK.append(Inferensi(fuzzyset[i]))

for i in range(0,100) :
  tabBengkel[i].Kelayakan = defuzzification(arrayNK[i].Rendah, arrayNK[i].Tinggi)

SL = tabBengkel

insertionSort(SL)

for i in range(0,100) :
  print(SL[i].ID,SL[i].Servis,SL[i].Harga,SL[i].Kelayakan)

```
